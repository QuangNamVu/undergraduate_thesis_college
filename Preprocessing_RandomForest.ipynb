```python
import os
import datetime
import pandas as pd
```

```python
home_path = os.path.expanduser("~")

data_file_name = home_path + "/data/ccxt/extra/BTC_USDT_binance_1h.csv"

output_file_name = home_path + "/data/ccxt/preprocessing_data/BTC_USDT_binance_1h_RF.csv"

attributes_normalize_mean = ['Open','High','Low','Close','Volume','N_buy','buy_amount_avg','sell_amount_avg','buy_amount_std','sell_amount_std','price_avg','cost_avg','cost_std','Spread_Open_Close','Spread_High_Low','MA_Close_24','MA_Close_240']

columns = ['Timestamp','Open','High','Low','Close','Volume','N_buy','N_sell','buy_amount_avg','buy_amount_std','sell_amount_std','price_avg','price_std','cost_avg','cost_std','MA_Close_24','MA_Close_240']
```

```python
df = pd.read_csv(data_file_name)
df = df[columns]
```

# Spreading Open-Close and High-Low

```python
df['Spread_Open_Close'] = df.Open - df.Close
df['Spread_High_Low'] = df.High - df.Low
df['Spread_Close_MA_Close_24'] = df.Close - df.MA_Close_24
df['Spread_Close_MA_Close_240'] = df.Close - df.MA_Close_240
```

```python
# columns.append('Spread_Open_Close')
# columns.append('Spread_High_Low')
columns =  [
    'Open',
    'High',
    'Low',
    'Close',
    'Volume',
    'N_buy',
    'N_sell',
    'buy_amount_avg',
    'buy_amount_std',
    'sell_amount_std',
    'price_avg',
    'price_std',
    'cost_avg',
    'cost_std',
    'MA_Close_24',
    'MA_Close_240',
    'Spread_Open_Close',
    'Spread_High_Low']
```

# 1 d diffirent

```python
_1d_cols = [
    'Open',
    'High',
    'Low',
    'Close',
    'Volume',
    'N_buy',
    'N_sell',
    'buy_amount_avg',
    'buy_amount_std',
    'sell_amount_std',
    'price_avg',
    'price_std',
    'cost_avg',
    'cost_std',
    'MA_Close_24',
    'MA_Close_240',
    'Spread_Open_Close',
    'Spread_High_Low']
```

```python
for att in _1d_cols:
    df['_1d_' + att] = df[att].diff(periods=1)
    
df
```

# Labeling

```python
df['1d_target'] = df.Close.shift(periods=-1) - df.Close
df = df.dropna()
df['trend'] = [1 if x >=0 else 0 for x in df['1d_target']]
df
```

```python
df.to_csv(output_file_name)
print(output_file_name)
```
