```python
import os
import datetime
import pandas as pd
```

```python
home_path = os.path.expanduser("~")

input_file_name = home_path + "/data/ccxt/raw/ETH_USDT_binance_1h.csv"

output_file_name = home_path + "/data/ccxt/preprocessing_data/ETH_USDT_binance_1h.csv"
```

```python
df = pd.read_csv(input_file_name)
df
```

```python
df['Unix_timestamp'] = df.Timestamp/1000.0
```

```python
pd.to_datetime(df['Unix_timestamp'],unit='s')
```

```python
df.Timestamp = pd.to_datetime(df['Unix_timestamp'],unit='s')
df
```

# Spreading Open-Close and High-Low

```python
df['Spread_Open_Close'] = df.Open - df.Close
df['Spread_High_Low'] = df.High - df.Low
df['Spread_Buy_Sell'] = df.buy_amount_avg * df.N_buy - df.sell_amount_avg * df.N_sell
df['Spread_Buy_Sell_avg'] = df.buy_amount_avg - df.sell_amount_avg
```

# MA

```python
# columns.append('Spread_Open_Close')
# columns.append('Spread_High_Low')
df['MA_Close_3'] = df['Close'].rolling(3).mean() # 3 hour
df['MA_Close_12'] = df['Close'].rolling(12).mean() # 12 hour
df['MA_Close_24'] = df['Close'].rolling(24).mean() # 1 day
df['MA_Close_120'] = df['Close'].rolling(120).mean() # 5 day
```

```python
_1d_cols = [
       'Open', 'High', 'Low', 'Close', 'Volume', 'N_buy',
       'N_sell', 'buy_amount_avg', 'sell_amount_avg', 'buy_amount_std',
       'sell_amount_std',
       'Spread_Open_Close', 'Spread_High_Low', 'Spread_Buy_Sell', 'Spread_Buy_Sell_avg', 'Spread_Buy_Sell_avg', 
       'MA_Close_3', 'MA_Close_12', 'MA_Close_24', 'MA_Close_120',
]
```

# 1 d diffirent

```python
for att in _1d_cols:
    df['_1d_' + att] = df[att].diff(periods=1)
    
df
```

# Labeling

```python
df['1d_target'] = df.Close.shift(periods=-1) - df.Close
df = df.dropna()
df['trend'] = [1 if x >=0 else 0 for x in df['1d_target']]
df
```

# Dump to file

```python
df.to_csv(output_file_name, index = None)
print(output_file_name)
```
