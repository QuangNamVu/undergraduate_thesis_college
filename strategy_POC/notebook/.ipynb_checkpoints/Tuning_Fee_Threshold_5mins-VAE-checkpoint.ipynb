```python
import matplotlib.pyplot as plt
import numpy as np
from decimal import Decimal
```

```python
# BTC fund amount
A_FUND_AMOUNT = Decimal('1.0')
# USDT fund amount
B_FUND_AMOUNT = Decimal('0.0')

FLAG_A_FUND_AMOUNT = Decimal('0.0')
FLAG_B_FUND_AMOUNT = Decimal('1.0')
FLAG_PREV_SIDE = 'buy'
# fee in real:
FREE_REAL = Decimal('.01')

A_LIMIT = Decimal('.000001')
B_LIMIT = Decimal('.01')

A_FUND_AMOUNT_LIST = []
B_FUND_AMOUNT_LIST = []
PRICE_LIST = []
```

```python
def buy():
    global PRICE, A_FUND_AMOUNT, B_FUND_AMOUNT, FREE_REAL
    # buy A sell B:
#     print("buy with price {}".format(PRICE))
    sell_amount_B = B_FUND_AMOUNT - (B_FUND_AMOUNT % B_LIMIT)
    
    cost = (sell_amount_B * FREE_REAL) / PRICE

    B_FUND_AMOUNT -= sell_amount_B
    A_FUND_AMOUNT += (sell_amount_B / PRICE - cost)
```

```python
def sell():
    global PRICE, A_FUND_AMOUNT, B_FUND_AMOUNT, FREE_REAL
    # buy B sell A:
#     print("sell with price {}".format(PRICE))
    sell_amount_A = A_FUND_AMOUNT - (A_FUND_AMOUNT % A_LIMIT)

    cost = sell_amount_A * FREE_REAL * PRICE

    A_FUND_AMOUNT -= sell_amount_A
    B_FUND_AMOUNT += (sell_amount_A * PRICE - cost)
```

```python
def strategy_01_current_price(threshold = 0.01):
    fee_assume = Decimal(threshold)
    global PRICE, A_FUND_AMOUNT, B_FUND_AMOUNT, Close_lst, Ts_lst
    global A_FUND_AMOUNT_LIST, B_FUND_AMOUNT_LIST, PRICE_LIST, SELL_TS, SELL_PRICE, BUY_TS, BUY_PRICE
    A_FUND_AMOUNT = Decimal(FLAG_A_FUND_AMOUNT)
    B_FUND_AMOUNT = Decimal(FLAG_B_FUND_AMOUNT)
    
    A_FUND_AMOUNT_LIST = []
    B_FUND_AMOUNT_LIST = []
    
    PRICE_LIST = []
    
    
    SELL_TS = []; BUY_TS = []; SELL_PRICE = []; BUY_PRICE = []
    PRICE = Decimal(str(Close_lst[0]))
    PREV_SIDE = FLAG_PREV_SIDE
    
    for NEW_PRICE, ts in zip(Close_lst[1:], Ts_lst[1:]):
        NEW_PRICE = Decimal(str(NEW_PRICE))
        if NEW_PRICE > PRICE / (Decimal('1.0') - fee_assume) and PREV_SIDE == 'buy':
            PRICE = NEW_PRICE
            sell()

            PREV_SIDE = 'sell'
            PRICE_LIST.append(PRICE)
            SELL_TS.append(ts)
            SELL_PRICE.append(PRICE)
            A_FUND_AMOUNT_LIST.append(A_FUND_AMOUNT)
            B_FUND_AMOUNT_LIST.append(B_FUND_AMOUNT)
            

        elif NEW_PRICE < PRICE * (Decimal('1.0') - fee_assume) and PREV_SIDE == 'sell':
            PRICE = NEW_PRICE
            buy()

            PREV_SIDE = 'buy'
            PRICE_LIST.append(PRICE)
            BUY_TS.append(ts)
            BUY_PRICE.append(PRICE)
            A_FUND_AMOUNT_LIST.append(A_FUND_AMOUNT)
            B_FUND_AMOUNT_LIST.append(B_FUND_AMOUNT)


    print("Number of trade: ", len(A_FUND_AMOUNT_LIST))

    print(A_FUND_AMOUNT, "  BTC")
    print(B_FUND_AMOUNT, "  USDT")

    print("Init price USDT is ", Decimal('1.0') *
          Decimal(Close_lst[0]), "  USDT")
    
    balance = A_FUND_AMOUNT * Decimal(Close_lst[-1]) + B_FUND_AMOUNT
    print("Last amount USDT is ", balance , "  USDT")
    return balance
```

# Simulate: If know trending of price in future

```python
def strategy_01_future_price(Predict_zip, threshold = 0.01):
    fee_assume = Decimal(threshold)
    global PRICE, A_FUND_AMOUNT, B_FUND_AMOUNT, Close_lst, Ts_lst
    global A_FUND_AMOUNT_LIST, B_FUND_AMOUNT_LIST, PRICE_LIST, SELL_TS, SELL_PRICE, BUY_TS, BUY_PRICE
    A_FUND_AMOUNT = Decimal(FLAG_A_FUND_AMOUNT)
    B_FUND_AMOUNT = Decimal(FLAG_B_FUND_AMOUNT)
    A_FUND_AMOUNT_LIST = []
    B_FUND_AMOUNT_LIST = []
    
    PRICE_LIST = []
    SELL_TS = []; BUY_TS = []; SELL_PRICE = []; BUY_PRICE = []
#     PRICE = Decimal(str(Close_lst[0]))
    PRICE = Decimal(str(Predict_zip[0][0]))
    PREV_SIDE = FLAG_PREV_SIDE
    for (NEW_PRICE, trend, ts) in Predict_zip:
        
#         assert 
        NEW_PRICE = Decimal(str(NEW_PRICE))
        if NEW_PRICE > PRICE / (Decimal('1.0') - fee_assume) and trend == 1 and PREV_SIDE == 'buy':
            PRICE = NEW_PRICE
            sell()

            PREV_SIDE = 'sell'
            PRICE_LIST.append(PRICE)
            SELL_TS.append(ts)
            SELL_PRICE.append(PRICE)
            A_FUND_AMOUNT_LIST.append(A_FUND_AMOUNT)
            B_FUND_AMOUNT_LIST.append(B_FUND_AMOUNT)
            

        elif NEW_PRICE < PRICE * (Decimal('1.0') - fee_assume) and trend == 0 and PREV_SIDE == 'sell':
            PRICE = NEW_PRICE
            buy()

            PREV_SIDE = 'buy'
            PRICE_LIST.append(PRICE)
            BUY_TS.append(ts)
            BUY_PRICE.append(PRICE)
            A_FUND_AMOUNT_LIST.append(A_FUND_AMOUNT)
            B_FUND_AMOUNT_LIST.append(B_FUND_AMOUNT)


    print("Number of trade: ", len(A_FUND_AMOUNT_LIST))

    print(A_FUND_AMOUNT, "  BTC")
    print(B_FUND_AMOUNT, "  USDT")

    
    print("Init price USDT is ", Decimal('1.0') *
          Decimal(str(Predict_zip[0][0])), "  USDT") # 1 BTC
    
    balance = A_FUND_AMOUNT * Decimal(Decimal(str(Predict_zip[-1][0]))) + B_FUND_AMOUNT
    print("Last amount USDT is ", balance , "  USDT")
    return balance
```

# VAE inference profit

```python
checkpoint_path = './train_log/train/checkpoint'
import VAE
```

```python
from tensorpack.predict.config import PredictConfig
from tensorpack.tfutils import SmartInit
```

```python
from VAE.Model import ModelDesc, Trainer  # , RandomZData
from VAE import get_default_hparams

hps = get_default_hparams()
M = ModelDesc(hps)
```

```python
pred_config = PredictConfig(
        session_init=SmartInit(checkpoint_path),
        model=M,
        input_names=['x'],
        output_names=['predict/y_pred']
)
```

```python
from tensorpack.predict import OfflinePredictor
predictor = OfflinePredictor(pred_config)
```

```python
N_val = X_val.shape[0]
X_val_vae = np.zeros([N_val - hps.T + 1, hps.T, hps.D])
y_val_vae = np.zeros([N_val - hps.T + 1])
ts_val_vae = np.zeros([N_val - hps.T + 1])

for i in range(N_val - hps.T + 1):
    
    X_val_vae[i] =  X_val[i : i + hps.T]    
    y_val_vae[i] = y_val[i + hps.T - 1]
    ts_val_vae[i] = ts_val[i + hps.T -1]
    
X_val_vae[-1].shape
```

```python
from VAE.load_data import test_segment

N_val = X_val.shape[0]

X_val_vae = test_segment(df[cols].values[-N_val:], hps)
N_val_vae = X_val_vae.shape[0]
y_val_vae = y_val[-N_val_vae:]
Close_vae = df.Close.values[-N_val_vae:]

print(N_val_vae)
```

```python
# tf.TensorSpec((None, self.hps.T, self.hps.D), tf.float32, 'x'),
M = hps.M

N_val = X_val_vae.shape[0]  #  25916
y_val_hat = np.zeros([N_val, 1, hps.C])
y_val_hat[-M:] = predictor(X_val_vae[-M:])[0]

for i in range(N_val // M):
    idx = M * i
    y_val_hat[M * i:M *(i + 1)] = predictor(X_val_vae[M * i : M *(i+ 1)])[0]
    
y_val_hat = np.argmax(y_val_hat[:,0,:], axis = -1)


if len(y_val_vae.shape) == 3:
    y_val_vae = np.argmax(y_val_vae[:,0,:], axis = -1)

y_val_hat
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
cm = confusion_matrix(y_val_vae, y_val_hat)
rp = classification_report(y_val_vae, y_val_hat)
score = accuracy_score(y_val_vae, y_val_hat)

print(cm)
print(rp)
print(score)
```
