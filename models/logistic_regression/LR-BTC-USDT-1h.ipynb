```python
import numpy as np
import os
import pandas as pd
home_path = os.path.expanduser("~") + '/'
from utils.load_data import *
from utils.info_params import *
```

```python
train_data_file_name = home_path + "data/ccxt/preprocessing_data/BTC_USDT_binance_1h.csv_train"
test_data_file_name = home_path + "data/ccxt/preprocessing_data/BTC_USDT_binance_1h.csv_test"
normalize_data = 'min-max'
normalize_data_idx = True
```

```python
df_train = pd.read_csv(train_data_file_name)
df_train
```

```python
df_test = pd.read_csv(test_data_file_name)
df_test
```

```python
df_train.columns
```

```python
cols = [
       'Open', 'High', 'Low', 'Close', 'Volume', 'N_buy',
       'N_sell', 'buy_amount_avg', 'sell_amount_avg', 'buy_amount_std',
       'sell_amount_std', 'price_avg', 'price_std', 'cost_avg', 'cost_std',
       'Spread_Open_Close', 'Spread_High_Low', 'Spread_Buy_Sell', 'Spread_Buy_Sell_avg',
       'MA_Close_3', 'MA_Close_12', 'MA_Close_24', 'MA_Close_120',
       '_1d_Open', '_1d_High', '_1d_Low', '_1d_Close', '_1d_Volume', '_1d_N_buy', '_1d_N_sell',
       '_1d_buy_amount_avg', '_1d_sell_amount_avg', '_1d_buy_amount_std',
       '_1d_sell_amount_std', '_1d_price_avg', '_1d_price_std', '_1d_cost_avg',
       '_1d_cost_std', '_1d_Spread_Open_Close', '_1d_Spread_High_Low',
       '_1d_Spread_Buy_Sell', '_1d_Spread_Buy_Sell_avg',
       '_1d_MA_Close_3', '_1d_MA_Close_12', '_1d_MA_Close_24', '_1d_MA_Close_120',
]
```

```python
N_val = 3 * 30 * 24 # 3 month = 2160 hours
N_val
```

```python
X = df_train[cols].values
y = df_train['trend'].values


X_train = X[:-N_val]
y_train = y[:-N_val]
X_val = X[-N_val:]
y_val = y[-N_val:]

X_test = df_test[cols].values
y_test = df_test['trend'].values
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(C = 0.1)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

y_train_hat = clf.predict(X_train)
cm = confusion_matrix(y_train, y_train_hat)
rp = classification_report(y_train, y_train_hat)
score = accuracy_score(y_train, y_train_hat)
f1 = f1_score(y_train, y_train_hat)

print(cm)
print(rp)
print(score)
print('f1_score : ', f1)
```

```python
y_pred = clf.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
rp = classification_report(y_val, y_pred)
score = accuracy_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)

print(cm)
print(rp)
print(score)
print('f1_score : ', f1)
```

# Test metrics

```python
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
rp = classification_report(y_test, y_pred)
score = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(cm)
print(rp)
print(score)
print('f1_score : ', f1)
```
